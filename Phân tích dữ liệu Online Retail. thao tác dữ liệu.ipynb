Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;

```python
temp2 = pd.to_datetime(df.loc[:,'date'],format='%Y-%m-%d')
df.loc[:,'Day']=temp2.dt.day
df.loc[:,'month']=temp2.dt.month
df.loc[:,'year']=temp2.dt.year
df.loc[:,['date','Day','month','year']]
```

df['Previous'] == (if df.month in [1,2,3]: 1,
                    elif df.month in [4,5,6]: 2, 
                      elif df.month in [7,8,9]:3,
                   elif df.month in [10,11,12]:4)

df.groupby(['month'])['Previous'].apply(if df.month in [1,2,3]: df.Previous.values == 1
                                         elif df.month in [4,5,6]: df.Previous.values == 2
                                        elif df.month in [7,8,9]: df.Previous.values == 3 
                                        elif df.month in [10,11,12]: df.Previous.values == 4 )

```python
Previous = []
for row in df['month']:
    if row in [1,2,3] :    Previous.append('1')
    elif row in [4,5,6]:   Previous.append('2')
    elif row in [7,8,9]:  Previous.append('3')
    else:           Previous.append('4')
df['Previous'] = Previous
```

Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

```python
# df['Quantity'].unique()
```

```python
df['UnitPrice'].unique()
```

```python
df['Amount'] = df['Quantity']*df['UnitPrice']
```

```python
df
```

Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

```python
conditions = [(df['Country'] == 'United Kingdom') & (df['Previous'] == '4'),
             (df['Country'] == 'France')]
choices = [10, 5]

df['Discount %'] = np.select(conditions, choices, default=0)
df 
```

```python
df
```

Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

```python
df['Total'] = df['Amount'] - (df['Discount %']/100)*df['Amount']
```
