```python
import pandas as pd
data = pd.read_csv('OnlineRetail.csv', header=0, encoding='ISO-8859-1')
print('*'*20, 'Nhap du lieu thanh cong', '*'*20)
display(data)
```

```python
data.info()
```

```python
#data.columns = ['Số hóa đơn', 'Mã hàng hóa', 'Mô tả', 'Số lượng', 'Ngày', 'Đơn giá', 'Mã KH', 'Quốc gia']


data.rename(columns = {'InvoiceNo': 'Số Hóa Đơn','StockCode':'Mã Hàng Hóa','Description':"Mô Tả",'Quantity':'Số Lượng', 'InvoiceDate':'Ngày',
                      'UnitPrice':'Đơn Giá','CustomerID':'Mã Khách Hàng', 'Country':'Quốc Gia'}, inplace = True)
data
```

```python
# Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

data2= data.loc[:,'Mô Tả':'Số Lượng']
data2
data2.to_csv('OnlineRetail1.csv')
```

```python
# Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
data3= data.loc[:999,:]
data3
data3.to_excel("OnlineRetail1.xlsx")  
```

```python
# Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5

data4 = data[data['Số Lượng'] > 10]
data4
data4.to_hdf('OnlineRetail1.h5', key='df', mode='w')  
```

```python
# Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
data5= data.loc[1000:1999,'Số Lượng':'Đơn Giá']
data5
data5.to_json(orient="index")
```

```python
# Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
data6 = data[data['Số Hóa Đơn'] == '536365']
data6
data6.to_html('OnlineRetail.html')
```
